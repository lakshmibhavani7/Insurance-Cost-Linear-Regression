```python
for i in col:
  if df[i].dtype == "object":
    df[i] = df[i].fillna(df[i].mode()[0])
  else:
    df[i]= df[i].fillna(df[i].mean())
```

```python
#outliers
for col in df.columns:
    if df[col].dtype != 'object':
     plt.boxplot(df[col])
     plt.title(col)
     plt.show()
```

```python
x = ["bmi","past_consultations","Hospital_expenditure", "NUmber_of_past_hospitalizations", "Anual_Salary"]
```

```python
for i in x:
  Q1= df[i].quantile(0.25)
  Q3 = df[i].quantile(0.75)
  IQR = Q3-Q1
  lowerfence = Q1-1.5*IQR
  upperfence = Q3+ 1.5 *IQR
  df = df[(df[i] >= lowerfence) & (df[i] <= upperfence)]
```

```python
for col in df.columns:
    if df[col].dtype != 'object':
     plt.boxplot(df[col])
     plt.title(col)
     plt.show()
```

```python
#model building

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns = ["charges"]) #--- we are removing chasrges as it was target variable
y = df["charges"]
```

```python
y
```

```python
# x = questions
# y= answers

# learning phase
# x-train
# y-train

# testing phase
# x-test

# after testing
# imagine the predictions are y_pred

# y-test is having actual answers

# r2 score as a teacher willcompare the y_pred and y_test
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size =0.80,random_state =42)
```

```python
x_train
```

```python
y_train
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_test
```

```python
result = pd.DataFrame(columns=["Actual value","predicted value"])
result["Actual value"] = y_test
result["predicted value"] = y_pred
result
```

```python
from sklearn.metrics import *
```

```python
result = r2_score(y_test,y_pred)
```
